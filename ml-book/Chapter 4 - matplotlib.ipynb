```python
%matplotlib inline
import matplotlib.pyplot as plt

rainfall = [17,9,16,3,21,7,8,4,6,21,4,1]
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

plt.bar(range(len(rainfall)), rainfall, align='center', color='orange')
plt.xticks(range(len(rainfall)), months, rotation='vertical')
plt.show()
```

```python
# Pie Charts
%matplotlib inline
import matplotlib.pyplot as plt

labels = ['Chrome','IE','Firefox','Edge','Safari','Sogou','Opera','Others']
marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19]
explode = [0,0,0.5,0,0.8,0,0,0]
colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# returns a tuple ( ('patches', 'texts', 'autotexts') )
pie = plt.pie(marketshare, explode=explode, labels=labels, colors=colors,
    autopct="%.1f%%", # string or function used to label the wedges with their numeric value
    shadow=True,
    startangle=45
)

plt.axis('equal')
plt.title('Web Browser Marketshare - 2018')
plt.legend(pie[0], labels, loc=2)
plt.show()

plt.savefig("webbrowsers.png", bbox_inches="tight")
```

```python
# Scatter Plots

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(
    [1,2,3,4],   # x
    [1,8,27,64], # y
    'bo'         # blue circle marker
)
plt.axis([0,4.5,0,70]) # xmin, xmax, ymin, ymax
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(1,4.5,0.1)

plt.plot(a, a**2, 'y^', a, a**3, 'bo', a, a**4, 'r--')

plt.axis([0, 4.5, 0, 70])
plt.show()
```

```python
# subplots

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(1,5,0.1)

plt.subplot(121) # 1 row, 2 cols, chart 1
plt.plot(
    [1,2,3,4,5],
    [1,8,27,64,125],
    'y^'
)

plt.subplot(122) # 1 row, 2 cols, chart 2
plt.plot(a, a**2, 'y^', a, a**3, 'bo', a, a**4, 'r--')
plt.axis([0, 4.5, 0, 70])
plt.show()

# or plt.subplot(1,2,13) to have more than 10 charts in a single figure

# 'scatter()' draws points without lines
# 'plot()' may draw lines depending on parameters
```

```python
# Seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('drivinglicense.csv')

g = sns.catplot(
    x='gender', y='license', col='group',
    data=data, kind='bar', ci=None, aspect=1.0
)

g.set_axis_labels('', 'Proportion with Driving license')
g.set_xticklabels(['Men', 'Women'])
g.set_titles('{col_var} {col_name}')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
g = sns.catplot(
    x="who", y="survived", col="class",
    data=titanic, kind="bar", ci=None, aspect=1
)

g.set_axis_labels("", "Survival Rate")
g.set_xticklabels(["Men", "Women", "Children"])
g.set_titles("{col_name} {col_var}")

plt.show()
```

```python
# lmplot

import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.lmplot(
    x='petal_width', y='petal_length', data=iris,
    hue='species', palette='Set1',
    fit_reg=False, scatter_kws={"s": 70}
)

ax = plt.gca()
ax.set_title("Plotting using the Iris dataset")

plt.show()
```

```python
# Swarm Plots

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style('whitegrid')

data = pd.read_csv('salary.csv')

sns.swarmplot(x='gender', y='salary', data=data)

ax = plt.gca()
ax.set_title("Salary distribution")

plt.show()
```
